```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler
```

```python
# Load dataset
wine = load_wine()
X = wine.data
y = wine.target

# Convert to DataFrame
df = pd.DataFrame(X, columns=wine.feature_names)

df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
linked = linkage(X_scaled, method='ward')

plt.figure(figsize=(10, 6))
dendrogram(linked)
plt.title("Dendrogram")
plt.xlabel("Samples")
plt.ylabel("Distance")
plt.show()
```

```python
model = AgglomerativeClustering(n_clusters=3)
clusters = model.fit_predict(X_scaled)
```

```python
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=clusters, cmap='viridis')
plt.title("Hierarchical Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
for k in [2, 3, 4, 5]:
    model = AgglomerativeClustering(n_clusters=k)
    labels = model.fit_predict(X_scaled)
    
    plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=labels)
    plt.title(f"Clusters = {k}")
    plt.show()
```

```python
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score, homogeneity_score, completeness_score
```

```python
dbscan = DBSCAN(eps=1.5, min_samples=5)
db_labels = dbscan.fit_predict(X_scaled)
```

```python
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=db_labels, cmap='plasma')
plt.title("DBSCAN Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
noise = db_labels == -1

plt.scatter(X_scaled[~noise, 0], X_scaled[~noise, 1], c=db_labels[~noise])
plt.scatter(X_scaled[noise, 0], X_scaled[noise, 1], color='red', label='Noise')
plt.legend()
plt.title("DBSCAN with Noise Highlighted")
plt.show()
```

```python
for eps in [0.5, 1.0, 1.5, 2.0]:
    dbscan = DBSCAN(eps=eps, min_samples=5)
    labels = dbscan.fit_predict(X_scaled)
    
    plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=labels)
    plt.title(f"eps = {eps}")
    plt.show()
```

```python

if len(set(db_labels)) > 1:
    sil_score = silhouette_score(X_scaled, db_labels)
else:
    sil_score = -1

hom_score = homogeneity_score(y, db_labels)
comp_score = completeness_score(y, db_labels)

print("Silhouette Score:", sil_score)
print("Homogeneity Score:", hom_score)
print("Completeness Score:", comp_score)
```
